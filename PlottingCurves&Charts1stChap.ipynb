### **Plotting a simple curve**

```python
import matplotlib.pyplot as plt
x = range(100)
y = [value**2 for value in x]
plt.plot(x,y)
plt.show()
```

### **plotting sin(x) curve**

```python
import math
```

```python
T = range(100)
x =[(2*math.pi*t) /len(T) for t in T]
y = [math.sin(value) for value in x]
plt.plot(x,y)
```

### **plotting sin-2 curve**

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(-3,2,200)
y = x**2 - 2*x+1
plt.plot(x,y)
```

### **sin(x) and cos(x) in the [0,2pi] interval as follows**

```python
x = np.linspace(0,2*np.pi,100)
ya = np.sin(x)
yb = np.cos(x)
plt.plot(x,ya)
plt.plot(x,yb)
```

### **Deferred rendering**

```python
def plot_slope(x,y):
    xs = x[1:]-x[:-1]
    ys = y[1:]-y[:-1]
    plt.plot(x[1:],ys/xs)
x = np.linspace(-3,3,100)
y = np.exp(-x**2)
plt.plot(x,y)
plot_slope(x,y)
```

### **plotting multiple bar charts**

```python
data = [[5., 25., 50.,20.],[4.,23.,51.,17.],
[6.,22.,52.,19.]]
x = np.arange(4)
plt.bar(x+0.00,data[0],color='b',width=0.25)
plt.bar(x+0.25,data[1],color='g',width=0.25)
plt.bar(x+0.50,data[2],color='r',width=0.25)
```

### **Plotting Stacked bar charts**

```python
a = [5., 30.,45.,22.]
b = [5.,25.,50.,20.]
x = range(4)
plt.bar(x,a, color = 'b')
plt.bar(x,b, color = 'r', bottom = a)
```

```python
A = np.array([5.,30.,45.,22.])
B = np.array([5.,25.,50.,20.])
C = np.array([1.,2.,1.,1.])
X = np.arange(4)
plt.bar(X,A, color = 'b')
plt.bar(X,B, color='r', bottom = A)
```

```python
A = np.array([5., 30.,45.,22.])
B = np.array([5., 25.,50.,20.])
C = np.array([1.,2.,1.,1.])
X = np.arange(4)
plt.bar(X, A, color = 'b')
plt.bar(X, B, color = 'g', bottom = A)
plt.bar(X, C, color = 'r', bottom = A+B)
```

### Plotting back to back bar charts

```python
women_pop = np.array([5.,30.,45.,22.])
men_pop = np.array([5.,25.,50.,20.])
x = np.arange(4)
plt.barh(x, women_pop, color = 'r')
plt.barh(x, -men_pop, color = 'b')
```
